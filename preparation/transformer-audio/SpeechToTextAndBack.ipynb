```python
import os
import time
import numpy as np
from typing import List, Dict, Tuple
from dataclasses import dataclass
import sounddevice as sd
import soundfile as sf
from scipy.io import wavfile
from gtts import gTTS
from IPython.display import Audio

# Для распознавания речи
import speech_recognition as sr

# Для синтеза речи
import pyttsx3

# Для разделения спикеров и классификации тем
from transformers import pipeline
from pyannote.audio import Pipeline

from IPython.display import Javascript, Audio

RECORD_JS = """
const record = (time) => new Promise((resolve) => {
  navigator.mediaDevices.getUserMedia({ audio: true })
    .then((stream) => {
      const recorder = new MediaRecorder(stream);
      const chunks = [];
      recorder.ondataavailable = (e) => chunks.push(e.data);
      recorder.start();
      setTimeout(() => {
        recorder.stop();
        stream.getTracks().forEach(track => track.stop());
      }, time * 1000);
      recorder.onstop = () => resolve(
        new Blob(chunks, { type: 'audio/wav' })
      );
    });
});
"""


@dataclass
class Utterance:
    text: str
    speaker: str
    topic: str
    timestamp: float

class SpeechProcessor:
    def __init__(self):
        # Инициализация компонентов
        self.recognizer = sr.Recognizer()

        # Загрузка моделей для анализа
        self._load_models()

        # Настройки
        self.sample_rate = 16000
        self.chunk_duration = 5  # длительность сегментов в секундах

    def _load_models(self):
        """Загрузка моделей для диаризации и классификации тем"""
        # Модель для разделения спикеров (требуется токен HuggingFace)
        self.diarization_pipeline = Pipeline.from_pretrained(
            "pyannote/speaker-diarization",
            use_auth_token="")

        # Модель для классификации тем
        self.topic_classifier = pipeline(
            "text-classification",
            model="facebook/bart-large-mnli",
            tokenizer="facebook/bart-large-mnli")

        # Предопределенные темы
        self.possible_topics = [
            "business", "entertainment", "technology",
            "science", "health", "politics", "sports"
        ]

    def record_audio(duration=5):
      display(Javascript(RECORD_JS))
      js_result = eval_js(f"record({duration})")
      return js_result

    def save_audio(self, audio: np.ndarray, filename: str):
        """Сохранение аудио в файл"""
        wavfile.write(filename, self.sample_rate, audio)
        print(f"Аудио сохранено как {filename}")

    def recognize_speech(self, audio_file: str = None, language: str = "ru-RU") -> str:
        """Распознавание речи из файла или с микрофона"""
        if audio_file:
            # Распознавание из файла
            with sr.AudioFile(audio_file) as source:
                audio = self.recognizer.record(source)

        try:
            text = self.recognizer.recognize_google(audio, language=language)
            print(f"Распознанный текст: {text}")
            return text
        except sr.UnknownValueError:
            print("Речь не распознана")
            return ""
        except sr.RequestError:
            print("Ошибка сервиса распознавания")
            return ""

    def text_to_speech(self, text: str, language: str = "ru"):
      """Исправленный синтез речи с gTTS"""
      try:
          tts = gTTS(text=text, lang=language, slow=False)
          tts.save("temp_audio.mp3")

      except Exception as e:
          print(f"Ошибка синтеза речи: {e}")

    def diarize_audio(self, audio_file: str) -> list:
        """Разделение аудио на сегменты по спикерам"""
        try:
            # Применяем модель диаризации
            diarization = self.diarization_pipeline(audio_file)

            # Формируем список сегментов (начало, конец, спикер)
            segments = []
            for turn, _, speaker in diarization.itertracks(yield_label=True):
                segments.append({
                    'start': turn.start,
                    'end': turn.end,
                    'speaker': speaker
                })
            return segments

        except Exception as e:
            print(f"Ошибка диаризации: {e}")
            return []

    def process_conversation(self, audio_file: str) -> List[Utterance]:
        """Полная обработка разговора: распознавание, диаризация, классификация"""
        # Шаг 1: Разделение по спикерам
        speaker_segments = self.diarize_audio(audio_file)

        utterances = []

        # Шаг 2: Обработка каждого сегмента
        for start, end, speaker in speaker_segments:
            # Извлекаем сегмент аудио
            audio, sr = sf.read(audio_file)
            segment = audio[int(start*sr):int(end*sr)]

            # Сохраняем временный файл для распознавания
            temp_file = "temp_segment.wav"
            sf.write(temp_file, segment, sr)

            # Распознаем речь
            text = self.recognize_speech(temp_file)

            if text:
                # Классифицируем тему
                topic = self.classify_topic(text)

                # Создаем объект высказывания
                utterance = Utterance(
                    text=text,
                    speaker=speaker,
                    topic=topic,
                    timestamp=start
                )
                utterances.append(utterance)

            # Удаляем временный файл
            if os.path.exists(temp_file):
                os.remove(temp_file)

        return utterances

    def real_time_processing(self, duration: int = 30):
        """Обработка аудио в реальном времени"""
        print(f"Начинаю обработку в реальном времени на {duration} секунд...")
        start_time = time.time()

        while time.time() - start_time < duration:
            # Записываем аудио сегмент
            audio = self.record_audio(duration=self.chunk_duration)

            # Сохраняем во временный файл
            temp_file = "temp_realtime.wav"
            self.save_audio(audio, temp_file)

            # Обрабатываем сегмент
            utterances = self.process_conversation(temp_file)

            # Выводим результаты
            for utt in utterances:
                print(f"\n[Спикер: {utt.speaker}]")
                print(f"Тема: {utt.topic}")
                print(f"Текст: {utt.text}")
                print("-" * 50)

            # Удаляем временный файл
            if os.path.exists(temp_file):
                os.remove(temp_file)

if __name__ == "__main__":
    processor = SpeechProcessor()

    # Демонстрация возможностей
    print("1. Тест распознавания речи с микрофона...")
    text = processor.recognize_speech("night_poem.wav")

    if text:
        print("\n2. Тест синтеза речи...")
        processor.text_to_speech(f"Вы сказали: {text}")
```
